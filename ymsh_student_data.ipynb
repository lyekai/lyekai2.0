```python
import pandas as pd
import plotly.express as px
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_excel('/content/drive/MyDrive/11111.xlsx')

df_fx = df[df['升學管道'] == '繁星推薦']

class_counts = df_fx.groupby('班級')['升學管道'].count()

fig = px.line(class_counts, x=class_counts.index, y='升學管道', title='各班級繁星推薦人數')
fig.update_layout(
    xaxis_title='班級',
    yaxis_title='人數'
)
fig.show()
#可看出升學管道為繁星推薦人數最多為319，當時是我們的語資班
#其次是301，為數資班，推測為資優班的校排%數較前，故得出一個合理結果
```
